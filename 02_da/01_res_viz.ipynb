# **Result visualization**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
RESULTS_ROOT = "../01_simulation/04_results/"
SEEDS = ['42', '1234', '1867', '613', '1001']
```

```python
p_data = pd.DataFrame()
#READING DATA:
for s in SEEDS:
    filename = RESULTS_ROOT+f'poccup_by_vehs_{s}.csv'
    pf = pd.read_csv(filename)
    pf["seed"] = [s]*len(pf)
    p_data = pd.concat([p_data, pf])
    
p_data["time"] = p_data["time"] - 4*24*60*60
p_data["time"] = p_data["time"].astype(int)
p_data["time_of_day"] = (p_data["time"] - (p_data["time"] // (24*60*60))*24*60*60) / (24*60*60) #converting to 0.0-1.0 and removing periodicity
```

```python
p_data
```

```python
parkings = p_data["parking_id"].unique()
```

## Values by parking lots

```python
p_data_by_parking = {}
for p in parkings:
    p_filt = p_data[p_data["parking_id"] == p]
    p_filt = p_filt.groupby("time_of_day")["occupancy"].mean()
    p_data_by_parking[p] = p_filt    
```

```python
for p in parkings:
    plt.plot(p_data_by_parking[p].index*24, p_data_by_parking[p].values)
    plt.title(p)
    plt.show()
```

```python
#VEH_ID = "h3408c1:0"
VEH_ID = "h482c1:0"
#VEH_ID = "h3120c1:0"
#VEH_ID = "h3271c2:0"
#VEH_ID = "carIn901:4"
#VEH_ID = "carIn106:4"
```

```python
p_vehs = p_data[p_data["veh_id"] == VEH_ID]
```

```python
p_vehs
```

```python
p_veh_by_parking = {}
for p in p_vehs["parking_id"].unique():
    p_filt = p_vehs[p_vehs["parking_id"] == p]
    p_filt = p_filt.groupby("time_of_day")["occupancy"].mean()
    p_veh_by_parking[p] = p_filt
```

```python
MIN_ALPHA = 0.1
MAX_ALPHA = 0.5
WINDOW = 300

bins = np.arange(0, 1, 1/(24*6)) #15 minutes resolution for histogram background

for p in p_veh_by_parking:
    #background:
    hist, bin_edges = np.histogram(p_veh_by_parking[p].index, bins=bins)
    color_min = min(hist)
    color_max = max(hist)
    has_legend = False
    for i in range(len(bins)-1):
        if hist[i]>0:
            label = "moving time" if not(has_legend) else ""
            has_legend = True
                
            plt.fill_between(x=[i*1/6, (i+1)*1/6], y1=0, y2=1.0,
                             alpha=(hist[i]-color_min)/(color_max-color_min)*(MAX_ALPHA-MIN_ALPHA)+MIN_ALPHA,
                             color="tab:orange",
                             label=label)
            
    #parking data:
    ma = p_data_by_parking[p].rolling(WINDOW).mean()
    std = p_data_by_parking[p].rolling(WINDOW).std()
    plt.plot(ma.index*24, ma.values, label="mean of occupancy")
    plt.fill_between(std.index*24, ma-std, ma+std, alpha=0.5, label="std of occupancy")
    #plt.plot(p_data_by_parking[p].index*24, p_data_by_parking[p].values)
    #plt.plot(p_veh_by_parking[p].index*24, p_veh_by_parking[p].values)
    plt.title(f"Occupancy of\n{p} parking lot")
    plt.xlabel("time [h]")
    plt.ylabel("occupancy")
    plt.xticks(ticks=np.arange(0, 26, 2))
    plt.legend()
    plt.show()
```
