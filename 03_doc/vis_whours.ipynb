# **Visualizing working hours**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
wh_df = pd.read_csv("working_hours.csv")
wh_df["time"] = wh_df["time"] / 3600
```

```python
opening_time = wh_df[wh_df["type"] == "opening"]
closing_time = wh_df[wh_df["type"] == "closing"]
```

```python
plt.figure(figsize=(10,4))
plt.grid(visible=True, axis="y", alpha=0.33)
plt.bar(opening_time["time"], opening_time["portion"], width=0.4, alpha=0.7, label="opening times")
plt.bar(closing_time["time"], closing_time["portion"], width=0.4, alpha=0.7, label="closing times")
plt.xlim(0,24)
plt.xticks(ticks=[0, 6, 7, 8, 9, 14, 16, 17, 18, 24])
plt.xlabel("time [h]")
plt.ylabel("proportion")
plt.legend(fontsize=9)
```
